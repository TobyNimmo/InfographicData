```python
import matplotlib.pyplot as plt
import pandas as pd

pay = pd.read_csv('gcse-results.csv', na_values = ['.',' '])
included=['Inner London','Outer London']
excluded=['Inner London','Outer London','North East','North West','Yorkshire and the Humber','East Midlands','West Midlands','East','London','South East','South West','England']
labels = []

for x in range(len(pay['Area'])):
    y=pay['Area'][x]
    if y in included:
        labels.append(y)

sizes = []

for x in range(len(pay['Attainment8'])):
    y=pay['Attainment8'][x]
    if pay['Area'][x] in included:
        if x == 0:
            sizes.append(0)
        else:
            sizes.append(y)

colors = ['yellow','green','blue','purple','red','orange']
patches, texts = plt.pie(sizes, colors=colors, shadow=True, startangle=90)
plt.legend(patches, labels, loc="upper right")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot
import pandas as pd

gcse = pd.read_csv('ks2-results.csv', na_values = ['.',' ']).sort_values('Area')
pay = pd.read_csv('ea-rate-and-er-by-eg-and-nation.csv', na_values = ['.',' ','!']).sort_values('Area')

excluded=['Inner London','Outer London','North East','North West','Yorkshire and the Humber','East Midlands','West Midlands','East','London','South East','South West','England']

x = []
y = []

for i in range(len(gcse['Mathematics - Percentage reaching the expected standard'])):
    temp=gcse['Mathematics - Percentage reaching the expected standard'][i]
    if not(gcse['Area'][i] in excluded):
            x.append(temp)
            
for i in range(len(pay['Employment percent'])):
    temp=pay['Employment percent'][i]
    if not(pay['Area'][i] in excluded):
        y.append(temp)

matplotlib.pyplot.scatter(x,y)

matplotlib.pyplot.show()
```
